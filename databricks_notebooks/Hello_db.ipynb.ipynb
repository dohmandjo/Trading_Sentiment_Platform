```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

spark.range(5).show()
```

```python
data = [(1, "AAPL", "10-K", "2024-02-01", "Sample filing text")]
columns = ["id", "company", "form_type", "filing_date", "text"]

df = spark.createDataFrame(data, columns)
display(df)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS sec_filings")
df.write.format("delta").mode("overwrite").saveAsTable("sec_filings.raw_filings")
```

```python
display(spark.sql("SELECT * FROM sec_filings.raw_filings"))
```
